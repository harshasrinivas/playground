```python
import urllib2
import string
from IPython.core.display import display, HTML
from scrapy.selector import Selector

def show_css(url, css_path):

    response = urllib2.urlopen(url)
    body = response.read()

    x = Selector(text=body).css(css_path).extract()
    x = [i.rstrip().lstrip() for i in x]
    x = [i for i in x if i != '']
    y = ['<span style="background-color: yellow;">' + i + '</span>' for i in x]
    
    new = body
    for i, j in zip(x, y):
        new = string.replace(new, i, j)
    display(HTML(new))
```

```python
show_css('https://scrapy.org/', '.navigation li::text')
```
